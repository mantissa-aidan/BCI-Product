```python
from scipy.io.wavfile import read
import matplotlib.pyplot as plt

# read audio samples
input_data = read("BYB_Recording_2019-01-18_18.15.16.wav")
audio = input_data[1]
# plot the first 1024 samples
plt.plot(audio[:])
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time")
# set the title  
plt.title("Sample Wav 1")
# display the plot
plt.show()
```
